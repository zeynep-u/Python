```python
df3 = df2.drop("date_of_birth", axis = 1)
```

```python
df3.info()
```

```python
# Kategorilerin sayısını hesaplayın
cins_sayisi = df3['breed'].value_counts()

# Sütun grafiği çiz
cins_sayisi.plot.bar()

plt.ylim(0, 1000)

# Grafiği göster
plt.show()
```

```python
df3["breed"]
```

```python
df3.dropna(inplace=True)
```

```python
df3.info()
```

```python
from sklearn.preprocessing import LabelEncoder

# Kategorik verileri sayısal verilere dönüştürme
label_encoder = LabelEncoder()
df3['animal_type'] = label_encoder.fit_transform(df3['animal_type'])
df3['breed'] = label_encoder.fit_transform(df3['breed'])
df3['color'] = label_encoder.fit_transform(df3['color'])
df3['sex_upon_outcome'] = label_encoder.fit_transform(df3['sex_upon_outcome'])
df3['age_upon_outcome'] = label_encoder.fit_transform(df3['age_upon_outcome'])
```
